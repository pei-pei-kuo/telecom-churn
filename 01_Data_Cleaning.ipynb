```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#read data
churnData = pd.read_excel('Data/Telco_customer_churn.xlsx')
demoData = pd.read_excel('Data/Telco_customer_churn_demographics.xlsx')
locData = pd.read_excel('Data/Telco_customer_churn_location.xlsx')
popData = pd.read_excel('Data/Telco_customer_churn_population.xlsx')
servData = pd.read_excel('Data/Telco_customer_churn_services.xlsx')
statusData = pd.read_excel('Data/Telco_customer_churn_status.xlsx')
```

```python
popData.head()
```

```python
rawData = churnData.merge(demoData,on='CustomerID').merge(locData,on='CustomerID').merge(servData,on='CustomerID').merge(statusData,on='CustomerID')
```

```python
rawData.info()
```

```python
rawData.drop(rawData.filter(regex='_y$').columns.tolist(),axis=1, inplace=True)
```

```python
rawData.columns = rawData.columns.str.replace(' ', '')
rawData.columns = rawData.columns.str.replace('_x', '')
```

```python
rawData.info()
```

```python
rawData =  rawData.drop(['Count','Country','State','LatLong','Latitude','Longitude',
                        'Partner','ChurnLabel','ChurnCategory','Quarter','CustomerStatus','ChurnCategory'], axis=1)
```

```python
reason = rawData.groupby(['ChurnReason'])\
    .agg({'ChurnReason': 'count'})
reason.rename(columns={"ChurnReason":'count'}).sort_values(by='count',ascending=False)
```

```python
reasontoCat={
"Attitude of support person":"attitude",
"Competitor offered higher download speeds":"service",
"Competitor offered more data":"service",
"Don't know":"others",
"Competitor made better offer":"price",
"Attitude of service provider":"attitude",
"Competitor had better devices":"service",
"Network reliability":"service",
"Product dissatisfaction":"service",
"Price too high":"price",
"Service dissatisfaction":"service",
"Lack of self-service on Website":"service",
"Extra data charges":"price",
"Moved":"others",
"Limited range of services":"others",
"Long distance charges":"price",
"Lack of affordable download/upload speed":"price",
"Poor expertise of phone support":"attitude",
"Poor expertise of online support":"attitude",
"Deceased":"others"
}
```

```python
rawData['ChurnReasonCategory'] = rawData['ChurnReason'].map(reasontoCat)
```

```python
rawData.info()
```

```python
reason = rawData.groupby(['ChurnReasonCategory','ChurnReason'])\
    .agg({'ChurnReason': 'count'})
```
